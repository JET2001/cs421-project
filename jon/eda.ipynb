```python
# Taken from Milton's Data processing
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

data = np.load("first_batch_multi_labels.npz")
X = pd.DataFrame(data["X"])
y = pd.DataFrame(data["y"])
yy = pd.DataFrame(data["yy"])

X = data["X"]
y = data["yy"]

XX = pd.DataFrame(X)
yy = pd.DataFrame(y)
XX.rename(columns={0: "user", 1: "item", 2: "rating"}, inplace=True)
yy.rename(columns={0: "user", 1: "label"}, inplace=True)

# Combine the two dataframes where the user is the same
merged_df = pd.merge(XX, yy, on="user")
merged_df.sample(50)
```

```python
# Checking rating distribution
plt.figure(figsize=(8, 5))
sns.countplot(x="rating", data=merged_df)
plt.title("Distribution of Ratings")
plt.show()

# Label distribution in yy
sns.countplot(x="label", data=merged_df)
plt.title("Distribution of User Labels")
plt.show()
```

```python
## Visualize 20 random users from class 0
import copy
import random
import os, shutil
def plot_figure(label):
    if os.path.exists(f"figs/label{label}"):
        shutil.rmtree(f"figs/label{label}")
        os.mkdir(f"figs/label{label}")
    else:
        os.mkdir(f"figs/label{label}")
        
    df = copy.deepcopy(merged_df)
    df = df[df.label == label]
    user_list = df.user.tolist()
    random.seed("normal")
    users = random.sample(user_list, 20)
    for i in range(20):
        user = users[i]
        user_data = df[df.user == user]
        display(user_data)
        user_item = user_data.item.tolist()
        user_rating = user_data.rating.tolist()
        
        plt.figure(figsize=(7,5))
        plt.title(f"Ratings for user ID {user}")
        plt.xlabel("Item ID")
        plt.ylabel("Rating")
        plt.scatter(user_item, user_rating)
        plt.savefig(f"figs/label{label}/label{label}-user-{user}.png")
        plt.show()
plot_figure(2)
```

```python
# create a new design matrix
from collections import namedtuple
import numpy as np
from joblib import Parallel, delayed
from tqdm import tqdm

DataSplit = namedtuple('DataSplit',['x_train', 'x_val', 'x_test', 'y_train', 'y_val', 'y_test'])

def construct_user_row(x_np, user_id)->np.array:
    '''
    For each user, we want an m * m matrix B such that B[i][j] = P(i = x[i] | j = x[j]). This means that B[i][i] will definitely be 1.
    Then we flatten this matrix into an 1 x m^2 row and use it as a row vector.
    '''
    print("user_id = ", user_id)
    m = [[0 for j in range(x_np.shape[1])] for i in range(x_np.shape[1])]
    cache = dict()
    # items = x.columns.tolist()
    # x_np = x.to_numpy() # convert df to numpy
    for i in range(x_np.shape[0]):
        # print("i = ", i)
        user_rating_i = x_np[user_id, i]
        condition_i = (x_np[:, i] == user_rating_i)
        count_i = np.sum(condition_i) + 1 # smoothing
        
        for j in range(x_np.shape[0]):
            user_rating_j = x_np[user_id, j]
            condition_ij =  np.logical_and(condition_i, x_np[:, j] == user_rating_j)
            count_ij = np.sum(condition_ij) + 1 # smoothing
            
            m[i][j] = np.log(count_ij) - np.log(count_i) # take log of probability
    m_np = np.array(m)
    return m_np.flatten().reshape(1, x_np.shape[1]**2)
    

def preprocess(split: DataSplit = None, x = None):
    import copy
    x_ = None
    if split is not None:
        x_ = split.x_train
    else:
        x_ = copy.deepcopy(x)
    if x_ is None:
        raise AssertionError("either data split or x must be defined")
    # frequency_counts = X.apply(pd.Series.value_counts).fillna(0)
    # user_list = X_train.index.tolist()
    new_mat = []
    x_np = x_.to_numpy()
    new_mat = Parallel(n_jobs = -1)(delayed(construct_user_row)(x_np, i) for i in tqdm(range(x_np.shape[0])))
    
    # for user_id in range(len(user_list)):
    #     print("Processing user_id", user_id)
    #     m = construct_user_row(X_train, user_id)
        
        # new_mat.append(copy.deepcopy(m))

    # Convert m to a matrix
    new_mat_np = np.vstack(new_mat)
    assert new_mat_np.shape == (x_np.shape[0], x_np.shape[1]**2)
    return new_mat_np
    # print("Frequency Counts")
    # display(frequency_counts)
    

split = DataSplit(X_train, X_val, X_test, y_train, y_val, y_test)
# new_mat_np = preprocess(split)
# new_mat_np
```
